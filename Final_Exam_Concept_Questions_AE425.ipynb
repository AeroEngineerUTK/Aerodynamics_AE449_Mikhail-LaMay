# Chapter 1 - Introduction
---

- What are the requirements for an aircraft to be in a trim state, what are the forces present, and how are those forces calculated?
    - The requirements to be in a trim state is “nose-up” trim. This is required in most aircraft to maintain steady level flight. Nose-up trim involves setting the tail surfaces to produce a greater down load on the aft portion of the fuselage, which adds to the wing loading and the total lift required from the wing if altitude is to be maintained.
    - Drag = Thrust
    - Weight = Lift
    - Drag = $\frac{1}{2} \rho v^2 S C_D$
---
- What determines an aircraft’s climb and descent rates?
    - Thrust & Drag determining an aircraft’s climb and descent rate. 
        - $T = D \rightarrow $ Aircraft is in Steady Level Flight
        - $T > D \rightarrow$ Aircraft Climbs
        - $T < D \rightarrow$ Aircraft Descends
    - Altitude Effects
        - At higher altitudes, density is lower, so airplanes must fly faster for the same weight and lift coefficient.
        - Lower density decreases available thrust from the engine.
---
- What factors affect an aircraft’s range and endurance, and how do we estimate these quantities?
    - Range
        - What determines an aircraft's range?
            - How fast fuel  is burned/energy consumed.
            - How much fuel the aircraft carries.
            - How fast the aircraft flies.
        - How do we estimate these quanities?
            - Range/Weight of fuel change per amount of fuel burned.
            - Cruise Climb & Altitude Gain Range equations
    - Endurance
        - Minimum fuel consumption for a time(t) occurs at the flight condition where the endurance factor is maximum.
---
- What is/are the underlying physics principle(s) of aerospace propulsion?
    - Pure definition to push/drive an object forward
        - In an aerospace system the way that thrust is generated
        - The act of propelling, the state of being propelled, a propelling force or impulse
    - Newtons 3rd Law
        - “To every action there is always opposed an equal reaction”
        - If the fluid exerts a force on the tube, then the tube exerts a force on the fluid
        - This results in a momentum change of the fluid
            - This is the basis of aerospace propulsion
        - Force (Thrust) – Net change in fluid momentum + Net pressure difference -or- Thrust is the net flux of total momentum through the engine
---
- What are the primary types of aerospace propulsion systems, how do they generate thrust, and how are their behaviors characterized?
    - Airbreathing
        - Includes the reciprocating, turbojet, turbofan, ramjet, turboprop, and turboshaft engines.
        - Requires air to mix into the combustion process
    - Non-Airbreathing
        - Includes rocket motors, nuclear propulsion systems, and electric propulsion systems.
        - Solid Fuel carry fuel with aircraft.
        - Momentum process
---
- What are thermal and propulsive efficiencies in the context of aircraft engines, and how are they defined?
    - Thermal Efficiency
        - Thermal Efficiency (n_T) is defined as the net rate of organized energy (shaft power of kinetic energy) out of the engine divided by the rate of thermal energy available from the fuel in the engine. The fuel’s available thermal energy is equal to the mass flow rate of the fuel (m^._f) times the fuel lower-heating value (h_PR).
    - Propulsive Efficiency
        - Propulsive Efficiency (n_P) of a propulsive system is a measure of how effectively the engine power (W^._OUT) is used to power the aircraft.
        - Propulsive Efficiency is the ratio of the aircraft power (thrust times velocity) to the power out of the engine (W^._OUT)
        - High Propulsive Efficiency requires the exit velocity to be approximately equal to the inlet velocity.
    - Thermal and Propulsive Efficiencies can be combined to give the overall efficiency (n_O) of a propulsive system. 


# Chapter 2 - Mechanics of Fluid Flow
---

- What is the difference between a control mass (surface) and a control volume?
    - Control Mass (Surface)
        - “Inner Boundary” or “Shell”
        - Boundary is the control mass
        - Hugs the hardware of the nozzle			
    - Control Volume
        - “Outer Boundary” or “Shrink Wrap”
        - Region is the control volume
        - “Shrink Wrapped” to the outside, but includes the inlet and outlet
---
- What are the conservation laws that govern fluid dynamics?
    - Conservation of Momentum
        -“Newtons 2nd Law Applied to a Fluid” –> F = ma
        - “The resultant of the forces acting on a particle is equal to the rate of change of the linear momentum of the particle.”
        - When two or more objects exert forces on each other, their total momentum always stays the same. This is called the law of conservation of momentum, and it helps to explain why aircraft and rockets move. A rocket engine sends out a highspeed jet of gas when it is fired.
    - Conservation of Mass (Continuity Equation)
        - The basis of the conservation of mass principle for fluid mechanics is that mass can neither be created nor destroyed within the volume or system of interest. If the inflow mass flow rate is not balanced by the outflow mass flow rate, then there will be a change in volume or density within the volume of interest.
---
- What is the first law of thermodynamics, and how is it used in fluid mechanics?
    - Conservation of Energy
        - The first law of thermodynamics describes the conservation of energy. It states that when heat is added to a system, it transforms to an equal amount of another form of energy. It does this by doing one or both of the following things:
            - Increasing the amount of internal energy within the system
            - Doing external work by leaving the system, thereby affecting the area around the system.
    - dE=δQ- δW
    - States that heat is a form of energy, and thermodynamic processes are therefore subject to the principle of conservation of energy. This mean that heat energy cannot be created or destroyed. 
    - States that energy can neither be created or destroyed; energy can only be transferred or changed from one form to another.
    - Conservation of energy is applied to fluid flow through Reynolds Transform Theorem (RTT)
        - RTT provides a general form for converting conservation laws from control mass to control volumes
        - Take arbitrary control volume (CV); can be moving…
        - Mass and energy (“heat” Q and Work “W”) can cross control surface (CS) boundaries
        - Forces also act on CS and mass in CV
        - Allows reference frame moving with control volume at constant velocity and no body forces.
        - Combines flow work and energy flux
---
- What is the second law of thermodynamics, and how is it used in fluid mechanics?
    - Conservation of Entropy
        - dS=δQ/T+δPs
        - The second law of thermodynamics states that no reaction is completely efficient, and that heat is always transferred from warmer objects to colder objects. In other words, some energy is always lost to heat in any reaction, and you will never find heat moving from colder objects to warmer objects in a system.
        - Entropy is a measure of how the energy is “spread out” in a process at a specific temperature
            - Also describes the ease that we access the energy
            - Also indicates which direction the energy will flow
            - For a reversible process
                - Ex: most ideal systems
---
- How are energy and enthalpy different, and how do we know which should be used?
    - Satisfying the 1st law does not guarantee a physically meaningful result
        - Ex: Electric resistance heating
    - Enthalpy is the measure of energy dispersal or spread at a specific temperature.
---
- What is required for a process to be isentropic?
    - Adiabatic & Reversible Processes
        - Adiabatic
            - No Heat Transfer
            - Relating to or denoting a process or condition in which heat does not enter or leave the system concerned.
    - Reversible Process
        - System and its’ surroundings return to their initial states if you undo the process (or complete a cycle)
        - P2/P1 = (T2/T1)^(gamma/(gamma-1))
        - P2/P1 = (row2/row1)^(gamma)
---
- What are the properties of ideal, calorically perfect gasses?
    - For calorically perfect gases, C_P and C_V are both constant
    - In AE425, we will model all gases as ideal and calorically perfect
        - This lets us say… C_P = C_V + R
---
- What are the various “iso-“ processes, and how are they plotted on T-s diagrams?
    - Isobaric Process  - Constant Pressure
    - Isothermal Process – Constant Temperature
    - Isochoric Process – Constant Volume/Density

---

- What are the governing equations of steady, 1D, compressible fluid flow?
    - $\frac{\partial \rho}{\partial t} + u \frac{\partial \rho}{\partial x} + \rho \frac{\partial u}{\partial x} = 0$
    - $\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} + \frac{1}{\rho} \frac{\partial P}{\partial x} = 0$
---
- How are the total and/or stagnation conditions defined for compressible flows?
    - Total Enthalpy
        - The total enthalpy is equal to the static enthalpy we would have if the flow were brought to rest adiabatically and without mechanical work
    - Total Temperature
        - The total temperature is not impacted by irreversibilities because it stems directly from a conservation law
    - Total Energy
        - Not conseved in the deceleration
    - Total Pressure
        - Defined as the pressure that would exist if the flow was brought to rest isentropically
        - Useful as measure of the flow's ability to do work in the engine
    - Total Density
        - Change Total Pressure variables to Total Density variables in its' equation
        - Also require isentropic deceleration
---
- How are pressure, Mach number, flow area related to mass-flow rate for both subsonic and supersonic flows?
    - Sonic Flow: "Choked Flow"
        - $\frac{\dot{m}}{A^* } = \frac{P_t}{\sqrt{T_t}} \sqrt{\frac{\gamma}{R}}[\frac{\gamma + 1}{2}]^{\frac{-(\gamma +1)}{2(\gamma - 1)}}$
    - Sub-Sonic Flow:
        - $\dot{m} = \frac{A p_t}{\sqrt{T_t}} \sqrt{\frac{\gamma}{R}} M (1 + \frac{\gamma-1}{2} M^2)^{- \frac{\gamma +1}{2(\gamma -1)}}$

# Chapter 4 - Shockwaves and Viscous Effects
---

- Shock Diamonds
    - Occurs when the flow is “over expanded”
    - Nozzle takes low K.E., high P flow and “expands” it to a lower pressure w/ higher K.E.
    - Over expansion – exit pressure is lower than ambient
        - Flow is trying to reach pressure equilibrium
        - Pressure equalizes through a shockwave
    - Shock wave reduces final K.E.
---
- Navier Stokes
    - Equations are non-linear PDEs
    - They are conservative
        - Support discontinuous solutions
            - Especially true in the inviscid limit
    - In steady flow, they require upstream supersonic flow – Circles are sound waves
    - Sudden rise in static pressure (irreversibility)
    - Sudden decrease in velocity with direction change
    - Sudden increase in entropy
---
- Normal Shock
    - 1D, constant area flow without heat addition
    - Flow through a normal shock is irreversible & adiabatic at constant $T_t$
    - Total Temp ($T_t$) = const
    - The ratios of the thermodynamic states are purely functioning of the Mach #.
    - Downstream Mach # is a unique function of upstream Mach #
    - The final conditions are known as the Rankine-Hugoniot jump conditions
        - Describes the relationship between the states on both sides of a shock wave or a combustion wave (deflagration or detonation) in a one-dimensional flow in fluids or a one-dimensional deformation in solids.
    - The conditions downstream of the shock  (ρ, P, T, u) are uniquely determined by those upstream of the shock.
    - Equations have a “symmetry” and are mathematically valid for both M_1>1 and M_2<1.
        - Are they physically valid for both?
    - Entropy jumps across the shock
        - Adiabatic, so $\Delta S \ge 0$ for physicality
    - $M_1 < 1$ , $M_2 > 1$ $\rightarrow \Delta S < 0 \rightarrow$ Non-Physcal Process
    - $M_1 = 1$ , $M_2 = 1$ $\rightarrow \Delta S = 0 \rightarrow$ Physical, but Trivial
    - $M_1 > 1$ , $M_2 < 1$ $\rightarrow \Delta S > 0 \rightarrow$ Physical Process
---
- 2nd Law of Thermodynamics (Normal Shock Wave)
    - Precludes the existence of "expansion shocks", even though continuity, momentum, and energy allow them in inviscid flows
    - Property Changes Across a Normal Shock Wave
        - Static Temp $\uparrow$
        - Static Pressure $\uparrow$
        - Mach # $\downarrow$
        - Density $\uparrow$
        - Velocity $\downarrow$
        - Total Temp const
        - Total Pressure $\downarrow$
        - Entropy $\uparrow$
---
- Oblique Shock Wave
    - Occur when supersonic flow is "turned into itself"
        - Geometric feature causes the flow to get in its' own way
    - Fluid gets deflected by an angle, $\theta$, through a shockwave at angle, $\beta$
    - For uniform upstream flow, the downstream flow will be uniform and parallel
    - Continuity, energy, and normal momentum equations are identical to those for normal shocks.
    - From tangential momentum, we find $u_{t_1}=u_{t_2}$
        - Shockwave can only change velocities normal to itself
        - Cannot change tangential velocity
    - Combination of oblique shock & isentropic flow are used to provide static temp ratio for higher efficiency.
    - In a supersonic flow inlet, the velocity is reduced from supersonic to subsonic speed with a normal shock.
        - The compression process is inefficient.
        - Several oblique shocks can be added to make the process more efficient
    - Overexpaned Flow
        - Pressure $\uparrow$
        - Speed $\downarrow$
        - Exit Condition Mach# > 1
        - Direction Changes

# Chapter 7 - Turbomachinery - Compressors
---

- Turbomachinery?
    - All devices in which energy is transferred to/from the fluid by the action of moving blade rows.
        - "Moving Parts"
        - Rotating blade rows change the enthalpy of the fluid
---
- What is the Euler Turbomachinery Equation, and what quantities does it relate?
    - $\dot{W_t} = \frac{\dot{m}\omega}{g_c}(r_i v_i - r_e v_e)$
    - $c_p(T_{ti}-T_{te}) = \frac{\omega}{g_c}(r_i v_i - r_e v_e)$
    -  The Euler Turbomachinery Equation relates the change in energy to the change in tangential momentum.
---
- How is the flow field through an axial compressor defined?
    - Throughflow $\rightarrow$ Describes axial and radial variations, blade rows are effectively thin actuator disks unifrom in $\theta$
    - Cascade Flow $\rightarrow$ Quasi-2D field tangential to the stream surfaces where the azimuthal direction is unwrapped to create an infinitely repeating domain
    - Secondary Flow $\rightarrow$ Viscous and threee dimentional flow
---
- What are the respective functions and resulting behaviors of the rotating and non-rotating blade in a compressor?
    - Inlet Guide Vanes (IGVs) $\rightarrow$ Stationary blade row that "pre-swirls" the flow for the first rotor
    - Rotor $\rightarrow$ Rotating blade row that increases angular velocity relative to the inertial frame. Changes $P_t$ & $T_t$
    - Stator $\rightarrow$ Stationary blade row that decreases the angular velocity (converts the KE to static pressure) to set up the flow for the next rotor
    - A stage is a combination of a rotor followed by a stator and is the fundamental unit of a compressor
---
- How does one construct a velocity triangle?
    - Draw rotor speed in direction rotor is spinning
    - Velocity vectors in the interial frame connect head-to-head with the rotor speed vector
    - The relative velocities connect head-to-tail with the rotor speed
---
- How do off-design conditions affect the flow behavior in a compressor?
    - If $\dot{m}$ changes $\rightarrow$ $u$ (axial velocity) changes
    - If $\Omega$ changes $\rightarrow$ $\Omega r$ will change
    - For either change, $\beta_1$ and $\alpha_2$ will change
---
- What are the limiting factors in axial-compressor performance?
    - Compressor works in an adverse pressure gradient
    - Seperation impacts pressure rise and efficiency
    - Rotating Stall and Surge may occure
---
- What is Surge (or Stall)?
    - Limit of steady operation for a compressor
        - Past the surge line, steady behavior is impossible
        - Pressure rise grows too large and the blades stall
        - Blade stall causes drop in pressure rise through stage
        - High pressure air in later stages can come back through the compressor
        - Adverse gradient lessens, and blade unstall
        - Process repeats unless operating conditions are changed
    - Unsteady loads may lead to blade failure
---
- Rotating Stall?
    - Departures from design $\frac{U}{\Omega r}$ changes local incidence angle, but not all blades will stall at the same time
        - The phenomenon rotates around the compressor disk
            - "Rotating Stall"
        - One or more cells of stalled flow embedded in an attached flow
            - Progresses around annulus at ~ 60% of the tagential velocity
---
- Design conciderations for best performance?
    - Compressor blades should operate at high Reynolds #
    - Turning angles must be limited due to adverse gradients
    - Large solidary tendes to reduce likelihood of seperation
        - Blade spacing
    - Subsonic relative Mach # help prevent shock-induced seperation
    - Blade design to handle seperation

# Chapter 8 - Turbomachinery - Turbines
---

- How do axial turbines compare to axial compressors in their operation and functionality?
    - The axial turbine is essentially the reverse of an axial compressor.
        - Removes energy from flow
        - Acts under a favorable pressure gradient
            - Greater turning angles
            - Greater pressure and temperature changes
            - Higher efficiency
    - In contrast to compressors, turbines operate under...
        - Higher mechanical stresses (higher loading)
        - Higher temperatures (right after the burner)
    - Turbines are simple devices - no different than a pinwheel
    - Turbine stage is a stator followed by a rotor
        - Stators are often referred to as "nozzles"
        - Rotors sometimes called buckets
---
- What is the general flow behavior in the stators and rotors of an axial turbine?
    - Nozzle
        - Flow is accelerated
        - Presssure decreases
        - Tagential velocity increases in direction of rotation
    - Rotor
        - Tangential velocity decreases
        - Resulting forces exert a force on the output shaft
        - Absolute/Inertial velocity magnitude decreases, while realtive velocity increases
---
- What is degree of reaction for a turbine stage? (Impulse Turbines - $R^\circ = 0$)
    -  No change in static pressure across the rotor
        - No change in total pressure in the blade-fixed frame (no work w/ no displacement)
        - Relative velocity magnitudes are unchanged
        - Velocity traiangle
---
- What are the behaviors of impulse and 50%-reaction turbines?
    - Enthalpy drop is equally balanced across the nozzle and rotor
    - Velocity triangles are symmetric top-bottom
        - $\alpha_2 = \beta_3$ & $\alpha_3  = \beta_2$ $\rightarrow$in terms of magnitude
    - Less torque than impulse turbine, but greater efficiency
        - $\alpha_3 = 0^\circ = \beta_2$ $\rightarrow$ $\Psi = 1$
    - Velocity triangles are right-angled
        - Flow coming in to the rotor is swirling at the same rate as the rotor
        - We have $\Delta U_\theta = \Omega r$
        - For the same $\Omega r_1$ the work of the 50% reaction stage is half that of the impulse stage for zero exit swirl
---
- What are the dominant stresses on a turbine rotor blade?
    - Thermal Stress
    - Bending Stress
    - Torsion Centrifugal Stress
    - Thermal Stress
    - Vibration
